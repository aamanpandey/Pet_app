```python
from datetime import date
import sys
class Pet():
    def __init__(self,name,dob,birth_weight,owner):
        self.name = name
        self.dob = dob
        self.birth_weight = birth_weight
        self.owner = owner
    
  
    #accessor methods
    def get_name(self):
        return self.name
    def get_dob(self):
        return self.dob
    def get_birth_weight(self):
        return self.birth_weight
    def get_owner(self):
        return self.owner
 
    #Mutator Methods
    def set_name(self, name):
        self.name = name
    def set_dob(self,dob):
        self.dob = dob
    def set_birth_weight(self,bw):
        self.birth_weight= bw
    def set_owner(self,owner):
        self.owner =owner
    
    def __str__(self):
        return 'Pet Name = ' + str(self.name) + '; Date of Birth = ' + str(self.dob) + '; Birth Weight = ' + str(self.birth_weight) + '; Owner = ' + str(self.owner)

class Mammels(Pet):
    def __init__(self,name,dob,birth_weight,owner,litter,has_claws):
        super().__init__(name,dob,birth_weight,owner)
        self.litter = litter
        self.has_claws = has_claws
        #self.calc_current_weight()
        
    #accessor methods
    def get_has_claws(self):
        return self.has_claws
    def get_litter(self):
        return self.litter
    
 
    #Mutator Methods
    def set_has_claws(self, has_claws):
        self.has_claws = has_claws
    def set_litter(self,litter):
        self.litter = litter
  
    
    def __str__(self):
        op=super().__str__()
        op+= 'Has Claws = ' + str(self.has_claws) + '; Litter = ' + str(self.litter) 
        return op
        
    def calc_current_weight(self):
        age_in_days = (date.today() - self.dob).days
        i = 0.08
        
        if age_in_days > 300:
            inc = 6
        else:
            inc = age_in_days // 50
      
        currentweight = self.birth_weight * ((1 + i) ** inc)
        
        return(currentweight)

class Fish(Pet):
    def __init__(self,name,dob,birth_weight,owner,scale_condition,length):
        super().__init__(name,dob,birth_weight,owner)
        self.scale_condition = scale_condition
        self.length = length
        
      #accessor methods
    def get_scale_condition(self):
        return self.scale_condition
    def get_length(self):
        return self.length
    
 
    #Mutator Methods
    def set_scale_condition(self, sc):
        self.scale_condition= sc
    def set_length(self,l):
        self.length = l   
        
        
    def calc_current_weight(self):
        agedays = (date.today() - self.dob).days
        i = 0.04
        
        if agedays > 360:
            inc = 4
        else:
            inc = agedays // 90
        
        currentweight = self.birth_weight * ((1 + i) ** inc)
        
        return(currentweight)
        
    def __str__(self):
        op=super().__str__()
        op+= 'Scale Condition = ' + str(self.scale_condition) + '; Length = ' + str(self.length) 
        return op
        
class Amphibian(Pet):
    def __init__(self,name,dob,birth_weight,owner,isVenemous):
        super().__init__(name,dob,birth_weight,owner)
        self.isVenemous = isVenemous
    
          #accessor methods
    def get_isVenemous(self):
        return self.isVenemous
 
    #Mutator Methods
    def set_scale_condition(self, iV):
        self.isVenemous= iV
 
    
    def calc_current_weight(self):
        agedays = (date.today() - self.dob).days
        
        if agedays//120<=3:
            currentweight = self.birth_weight * ((1 + 0.05) ** (agedays // 120))
        elif agedays > 360:
            currentweight = (self.birth_weight * ((1 + 0.05) ** (360 // 120)))+((self.birth_weight * ((1 + 0.03) ** ((agedays-360) // 120))))      
        return currentweight
    
    def __str__(self):
        op=super().__str__()
        op+= 'Is Venemous = ' + str(self.isVenemous)
        return op
    
class Person():
    def __init__(self,person_name,person_add):
        self.person_name = person_name
        self.person_add = person_add
    
    def __str__(self):  
        return 'Person Name = ' + str(self.person_name) + '; Person Address = ' + str(self.person_add)
```

```python
class WrongChoiceError(Exception):
    def __str__(self):
        return ("The input is not available")
    
class WrongDateError(Exception):    
    def __str__(self):
        return("The entered date is not in proper format")

```

```python
import pickle
class driver():
    #def pet_add():
        
    def menu_option(self):
        print('==== Menu ====')
        print('1. To add a pet')
        print('2. To print current weight of all pets')
        print('3. To print all pets and owners')
        print('4. To exit program')
        
        result=int(input('Your Selection: '))
        print()
        if (result not in (1,2,3,4)):
            raise WrongChoiceError()
        return result
                
    def get_date(self):
        date_entry = input("Enter the pet's date of birth (year,month,date): ")
        year,month,day = map(int,date_entry.split(','))
        if (len(str(year)) != 2 ):
            raise ValueError("Not a proper year format")    
        if (month <1 or month>12):
            raise WrongDateError()
        if (month in (1,3,5,7,8,10,12)):
            if (day <1 and day >31):
                raise WrongDateError()
        if (month in (2,4,6,9,11)):
            if (day <1 and day >30):
                raise WrongDateError()
        year = int(str('20')+str(year))
        #Adding 0 to the front if the month has single digit
        if (len(str(month)) == 1):
            month = int(str('0')+str(month))
        #Adding 1 to the front if day has single digit
        if (len(str(day)) == 1):
            day = int(str('0')+str(day))
        dob = date(year,month,day)
        if ((date.today() - dob).days <0):
            raise WrongDateError()
        return dob 
    
    def add_data(self):
        try:
            d=int(input('Type of Pet?(1-Mammal; 2-Fish, 3-Amphibian): '))
            if (d not in (1,2,3)):
                raise WrongChoiceError()
        except WrongChoiceError as error:
            print(str(error))
            print("Please enter the choice among the available ones")
            sys.exit("Unknown Error. Try again later")
            
        if (d >0 and d<4):
            if (d==1):
                name= input("Enter the pet's name: ")
                try:
                    dob=self.get_date()
                except ValueError as error:   
                    print("Error",error)
                    sys.exit()
                except WrongDateError as exception:
                    print(str(error))
                    print("Please enter the date in correct format")
                birth_weight= float(input("Enter the pet's birth weight: "))
                owner= input("Enter the owner's name: ")
                person_address= input("Enter the owner's address: ")
                litter=int(input('Enter the Litter Size: '))
                has_claws= bool(input("Does the pet has claws? Answer 'True' or 'False': "))
                #ab=mammal(name,dob,birth_weight,owner,litter,has_claws)
                with open('petdata.dat', 'ab') as fb:
                    ab_p = Person(owner,person_address)
                    ab_m = Mammels(name,dob,birth_weight,ab_p,litter,has_claws)
                    pickle.dump(ab_m, fb)
            if(d==2):
                name = str(input("Enter the pet's name: "))
                try:
                    dob = self.get_date()
                except ValueError as error:   
                    print("Error",error)
                except WrongDateError as exception:
                    print(str(error))
                    print("Please enter the date in correct format")
                weight = float(input("Please enter the pet's birth weight: "))
                owner = str(input("Please enter the owner's name: "))
                person_address = str(input("Please enter the owner's address: "))
                scale_cond = str(input("Please enter the scale condition: "))
                fish_length = float(input("Please enter the length of the fish: "))

                #Creating a person and mammals object object
                with open('petdata.dat','ab') as fb:
                    ab_p = Person(owner,person_address)
                    ab_f = Fish(name,dob,weight,ab_p,scale_cond,fish_length)
                    pickle.dump(ab_f,fb)
                    
            if(d == 3):
                name = str(input("Enter the pet's name: "))
                try:
                    dob = self.get_date()
                except ValueError as error:   
                    print("Error",error)
                except WrongDateError as exception:
                    print(str(error))
                    print("Please enter the date in correct format")
                weight = float(input("Please enter the pet's birth weight: "))
                owner = str(input("Please enter the owner's name: "))
                owner_address = str(input("Please enter the owner's address: "))
                isVenomous = bool(input("Is the pet venomous? Answer 'True' or 'False' :"))

                #Creating a person and mammals object object
                with open('petdata.dat','ab') as fb:
                    ab_p = Person(owner,owner_address)
                    ab_a = Amphibian(name,dob,weight,ab_p,isVenomous)
                    pickle.dump(ab_a,fb)
        
    def main(self):
        r=0
        while r!=4:
            try:
                r=self.menu_option()
            except WrongChoiceError as error:
                print(str(error))
                print("Please enter the choice among the available ones")
                sys.exit("Unknown Error. Try again later")
                
            if (r==1):
                self.add_data()
            if (r==2):
                d = []
                with open("petdata.dat", "rb") as fb:
                    while True: 
                        try:
                            d.append(pickle.load(fb))
                        except EOFError:
                            break
                print("Pet name and current weight of all Pets")
                print('*'*50)
                for ob in d:
                    #ob.calc_current_weight()
                    print(ob.name," weighs ",round(ob.calc_current_weight(),2)," pounds and is ",(date.today() - ob.dob).days," days old.")
            
            if (r==3):
                with open("petdata.dat", "rb") as f:
                    data=[]
                    while (True): 
                        try:
                            obj= pickle.load(f)
                            data.append(obj)

                        except EOFError:
                            break
                for obj in data:
                    if (isinstance(obj,Mammels) == True):
                        print("Details of this Mammal are")
                        print("Pet Name: ",obj.name,"; Pet DOB: ",obj.dob,"; Birth Weight: ",obj.birth_weight)
                        print("Owner Name: ",obj.owner.person_name,"; Owner Address: ",obj.owner.person_add)
                        print("Litter Size: ",obj.litter,"; Has Claws?",obj.has_claws)
                        print(" \n")
                    elif (isinstance(obj,Fish) == True):
                        print("Details of this Fish are")
                        print("Pet Name: ",obj.name,"; Pet DOB: ",obj.dob,"; Birth Weight: ",obj.birth_weight)
                        print("Owner Name: ",obj.owner.person_name,"; Owner Address: ",obj.owner.person_add)
                        print("Scale Condition: ",obj.scale_condition,": Length: ",obj.length)
                        print(" \n")
                    else:
                        print("Details of this Amphibian are")
                        print("Pet Name: ",obj.name,"; Pet DOB: ",obj.dob,"; Birth Weight: ",obj.birth_weight)
                        print("Owner Name: ",obj.owner.person_name,"; Owner Address: ",obj.owner.person_add)
                        print("Is Venomous: ",obj.isVenemous)
                        print(" \n")
                        
            if(r==4):
                print("You chose to exit the program")
                a = str(input("Are you sure (Y/N)?"))
                if (a == 'y' or a== 'Y'):
                    print("Data saved to petdata.dat")
                    print('Goodbye')
                    sys.exit(3)
                else:
                    print("Please Continue")
            
```
